## Part 2: Use KMeans to Assign Users to Clusters
We will use KMeans to associate every user to a cluster based on demographics. In a real world scenario we would then save that user/cluser association in a database. Here we will just reference the data in a local dataframe. 

We will then deploy the KMeans model to Watson ML. When a new user registers we will query Watson ML to determine the user's cluster. Again, in a real world scenario we would save the user/cluster association in the datbase.

The recommendation engine will be based on a user's cluster. We will look up the user's cluster (in the database or local dataframe) then popular products purchased by other users in the cluster.

### Clean and filter the Data
Convert age to a number, return only the features we'll need to create our clusters, and only return where age is not null.

```python
feature_cols = ['CUST_ID','GenderCode','AGE_CLEAN','NATIONALITY']
product_cols = ['Baby Food', 'Diapers', 'Formula', 'Lotion', 'Baby wash', 'Wipes', 'Fresh Fruits', 'Fresh Vegetables', 'Beer', 'Wine', 'Club Soda', 'Sports Drink', 'Chips', 'Popcorn', 'Oatmeal', 'Medicines', 'Canned Foods', 'Cigarettes', 'Cheese', 'Cleaning Products', 'Condiments', 'Frozen Foods', 'Kitchen Items', 'Meat', 'Office Supplies', 'Personal Care', 'Pet Supplies', 'Sea Food', 'Spices']
```

```python
df_filtered = df.select(['CUST_ID','GenderCode', df.AGE.cast('float').alias('AGE_CLEAN'), 'NATIONALITY']+product_cols).filter('AGE_CLEAN IS NOT NULL')
```

```python
display(df_filtered)
```

#### Build the KMeans Model
Here we import the required Spark ML libraries to build our model.

```python
from pyspark.ml import Pipeline
from pyspark.ml.clustering import KMeans
from pyspark.ml.clustering import KMeansModel
from pyspark.ml.feature import StringIndexer
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.linalg import Vectors
```

#### K-Means Implementation
Here we use K-Means to create 100 clusters based on the gender, age, and nationality of the users. 

```python
# Spark ML requires every feature to be an integer
# GenderCode and NATIONALITY are strings
# Here we use StringIndexer to convert them to integers
indexer_gender = StringIndexer(inputCol="GenderCode", outputCol="GENDER_IX")
indexer_nationality = StringIndexer(inputCol="NATIONALITY", outputCol="NATIONALITY_IX")
```

```python
assembler = VectorAssembler(inputCols=['GENDER_IX','AGE_CLEAN','NATIONALITY_IX'], outputCol='features')
```

```python
kmeans = KMeans()\
    .setK(100)\
    .setSeed(1)\
    .setFeaturesCol("features")\
    .setPredictionCol("prediction")
pipeline = Pipeline(stages=[indexer_gender, indexer_nationality, assembler, kmeans])
model = pipeline.fit(df_filtered)
```

#### Save Cluster for Each User 
Here we run the original dataset against the model to find the cluster for each user. 

```python
df_predictions = model.transform(df_filtered).select(['CUST_ID','prediction'])
df_predictions = df_predictions.groupby('CUST_ID','prediction').count().drop('count')
df_predictions = df_predictions.rdd.toDF() # Drops null columns OR forces execution sooner (instead of lazy transformation)
display(df_predictions)
```

#### Normalize the Dataset
Each row in our dataset represents a single checkout for a user, and each product purchased during that checkout has a value of 1. We want to get each product purchased by a user, and the total number of times that product has been purchased by the user, so we can make sure we are not recommending those products to the user. Here we create a new dataframe where each record contains a user, the product they purchased, and the number of times they have purchased that product. For example, we convert:

<pre>
User      Diapers      Milk
----      -------      ----
User A    0            1
User B    1            1
User B    0            1
User C    0            1
</pre>

to:

<pre>
User      Product      Count
----      -------      -----
User A    Milk         1
User B    Milk         2
User B    Diapers      1
User C    Milk         1
</pre>

```python
# This function takes a row in our dataframe and returns 0 or more rows - 1 for each cust/product purchase
def normalize(x, product_col_offset):
    l = []
    for i in range(product_col_offset,len(product_cols)+product_col_offset):
        if x[i] == 1:
            t = []
            for j in range(0, product_col_offset):
                t.append(x[j])
            t.append(product_cols[i-product_col_offset])
            t.append(1)
            l.append(tuple(t))
    return l
```

```python
# Here we convert the dataframe to an rdd and run a flatMap transformation
rdd_normalized = df_filtered.select(feature_cols+product_cols).rdd.flatMap(lambda x: normalize(x, len(feature_cols)))
df_normalized = rdd_normalized.toDF(feature_cols+['PRODUCT','COUNT'])
df_normalized = df_normalized.groupby(feature_cols+['PRODUCT']).sum('COUNT')
df_normalized = df_normalized.rdd.toDF(feature_cols+['PRODUCT','COUNT']) # Rename sum(COUNT) column back to COUNT
# join to add cluster
df_customer_product_cluster = df_normalized.alias('n').join(df_predictions.alias('p'), df_normalized['CUST_ID'] == df_predictions['CUST_ID']).select(['n.CUST_ID']+feature_cols[1:]+['p.prediction','n.PRODUCT','n.COUNT'])
display(df_customer_product_cluster)
```

### The Final Dataset
The final output above a dataframe called **df_customer_product_cluster**. This dataframe represents the database we would have available to us when it comes time to perform product recommendations. The next step is to deploy the model to Watson ML, so we can calculate the cluster for new users.

## Part 4: Product Recommendations
Let's create some product recommendations for our customer. We already now our customer's cluster, so here we find other popular products purchased by users in the cluster. We could cross-reference those against products the user has already purchased or are currently in their cart, so we don't recommend the same products. In a real world scenario this would be done in a backend application against a databse. Here we are treating the **df_customer_product_cluster** dataframe as our database.

```python
def get_recommendations(cust_prod_cluster_df, customer):
    cluster = customer[0].prediction
    # Existing customer products
    customer_products = cust_prod_cluster_df.filter('CUST_ID = {}'.format(customer[0].CUST_ID)).select('PRODUCT')
    print('PRODUCTS ALREADY PURCHASED:')
    customer_products.show()
    # Get popular products in the cluster
    cluster_products = cust_prod_cluster_df.filter('prediction = {}'.format(cluster))
    cluster_products = cluster_products.groupby('PRODUCT').sum('COUNT').sort(desc('sum(COUNT)'))
    print('POPULAR PRODUCTS IN CLUSTER:')
    cluster_products.show()
    # Filter out products the user has already purchased
    print('RECOMMENDED PRODUCTS:')
    recommended_products = cluster_products.alias('cl').join(customer_products.alias('cu'), cluster_products['PRODUCT'] == customer_products['PRODUCT'], 'leftouter')
    recommended_products = recommended_products.filter('cu.PRODUCT IS NULL').select('cl.PRODUCT','sum(COUNT)').sort(desc('sum(COUNT)'))
    recommended_products.show()
```

```python
get_recommendations(df_customer_product_cluster, customer)
```
